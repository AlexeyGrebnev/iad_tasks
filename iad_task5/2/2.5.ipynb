```python
import numpy as np

data = np.loadtxt('scipy-lectures.txt')
year, hares, lynxes, carrots = data.T  # колонки в переменные

import matplotlib.pyplot as plt
plt.axes([0.2, 0.1, 0.5, 0.8]) 

plt.plot(year, hares, year, lynxes, year, carrots) 

plt.legend(('Hare', 'Lynx', 'Carrot'), loc=(1.05, 0.5)) 

plt.show()
```

```python
print(f"Среднее отклонение для зайцев:\t {np.mean(hares)}")
print(f"Среднее отклонение для рысей:\t {np.mean(lynxes)}")
print(f"Среднее отклонение для морковки: {np.mean(carrots)}")
```

```python
print(f"Стандартное отклонение для зайцев:\t {np.std(hares)}")
print(f"Стандартное отклонение для рысей:\t {np.std(lynxes)}")
print(f"Стандартное отклонение для морковки:\t {np.std(carrots)}")
```

```python
print(f"Популяция вида была наибольшей для зайцев:\t {int(year[np.argmax(hares)])}")
print(f"Популяция вида была наибольшей для рысей:\t {int(year[np.argmax(lynxes)])}")
print(f"Популяция вида была наибольшей для морковки:\t {int(year[np.argmax(carrots)])}")
```

```python
max_species_each_year = np.array(['H', 'L', 'C'])[np.argmax(data[:, 1:], axis=1)]
print(f"Вид с наибольшей популяцией за каждый год: {max_species_each_year}")
```

```python
years_above_50000 = year[np.any(data[:, 1:] > 50000, axis=1)]
print(f"Годы, когда популяции превышают 50000: {years_above_50000}")
```

```python
print(f"Анти-Топ 2 годов для зайцев:\t {year[np.argsort(hares)[:2]]}")
print(f"Анти-Топ 2 годов для рысей:\t {year[np.argsort(lynxes)[:2]]}")
print(f"Анти-Топ 2 годов для морковки:\t {year[np.argsort(carrots)[:2]]}")
```

```python
data_table = np.loadtxt('https://scipy-lectures.org/_downloads/populations.txt')

years = data_table[:, 0]
hare = data_table[:, 1]
lynx = data_table[:, 2]

hares_grad= np.gradient(hare, years)
lynxes_grad = np.gradient(lynx, years)


plt.figure(figsize=(10, 6))
plt.plot(years, hares_grad, label='Изменение численности Зайцев')
plt.plot(years, lynxes_grad, label='Изменение численности рысей')


plt.title('График изменнения популяций Зайцев и рысей из файла')
plt.xlabel('Годы')
plt.ylabel('Значениe')
plt.legend()
plt.show()
```
