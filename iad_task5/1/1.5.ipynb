```python
import scipy
import requests
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
img_url="https://sun9-6.userapi.com/c4280/u8321716/-6/x_f53e1ba2.jpg"
r = requests.get(img_url, allow_redirects=True)
imgname="img1.jpg"
open(imgname, 'wb').write(r.content)
img = mpimg.imread(imgname)
plt.imshow(img)
plt.title('енотик')
plt.show()
```

```python
plt.imshow(img[:, :, 0])
plt.title('Цветовая карта енотика')
plt.show()  
```

```python
import numpy as np
from PIL import Image
img = np.array(Image.open(imgname))
sy, sx, sz = img.shape
y, x = np.ogrid[0:sy, 0:sx] # x и y обозначают пиксели
y.shape, x.shape
a=300; b=100

centerx, centery = (sx/2, sy/2) # центр изображения
mask = ((((y - centery-60)**2)/b**2) + (((x - centerx)**2)/a**2)) > 1 # круг
img[mask] = 60
plt.title('Глазки енотика (Овал - Эллипс)')
plt.imshow(img) 
plt.show()
```
